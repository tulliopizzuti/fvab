# Grafici


```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import os
```

```python
path= "C:\\Users\\angel\\OneDrive\\Desktop\\progettoBiometria\\hmog_dataset\\public_dataset"
meanSensors="MeanSensors.csv"
df= pd.read_csv(os.path.join(path, meanSensors))
featuresAccelerometer= df[["XAcc", "YAcc", "ZAcc"]]
featuresGyroscope= df[["XGyro", "YGyro", "ZGyro"]]
featuresMagnetometer= df[["XMagn","YMagn","ZMagn"]]



```

```python
fig, ax= plt.subplots(3,3)
ax[0,0].set_title("Grafico media Acceletrometro")
ax[0,0].plot(featuresAccelerometer,)
ax[1,1].set_title("Grafico media Gyroscope")
ax[1,1].plot(featuresGyroscope)
ax[2,2].set_title("Grafico media Magnetometer")
ax[2,2].plot(featuresMagnetometer)
```
