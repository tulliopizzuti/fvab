# Grafici


```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from os.path import join
```

Viene specificato il percorso della cartella principale contenenti i file "MeanSensorsSit.csv" e "MeanSensorsWalk.csv".

```python
path= "C:\\Users\\angel\\OneDrive\\Desktop\\progettoBiometria\\hmog_dataset\\public_dataset"
meanSensorsSitPath=join(path,"MeanSensorsSit.csv")
meanSensorsWalkPath=join(path,"MeanSensorsWalk.csv")
```

Apertura file csv

```python
csvMeanSensorsSit=pd.read_csv(meanSensorsSitPath)
csvMeanSensorsWalk=pd.read_csv(meanSensorsWalkPath)
columns = ["GestureScenario","XAcc", "YAcc", "ZAcc", "XGyro","YGyro", "ZGyro","XMagn","YMagn", "ZMagn"]
csvMeanSensorsSit.columns= columns 
csvMeanSensorsWalk.columns= columns
```

Creazione dei subplot della media per le variabili X, Y, Z dei tre sensori: Accelerometro, Giroscopio e Magnetometro. 
Viene mostrato il grafo per ogni coordinata X, Y, Z della media sovrapponendo i valori rispetto ai due scenari Sit e Walk.
Nell'ultimo grafico (per una questione di visibilit√†) viene rappresentata la legenda dove il colore verde rappresenta i dati per Sit, mentre il colore blu rappresenta i dati per Walk.


```python
fig, ax= plt.subplots(nrows=3, ncols=3,figsize=(15,15))
ax[0,0].set_title("X Accelerometro")
ax[0,0].plot(csvMeanSensorsSit.XAcc.values, color="green", alpha=0.5)
ax[0,0].plot(csvMeanSensorsWalk.XAcc.values, color="blue", alpha=0.5)
ax[0,1].set_title("Y Accelerometro")
ax[0,1].plot(csvMeanSensorsSit.YAcc.values, color="green", alpha=0.5)
ax[0,1].plot(csvMeanSensorsWalk.YAcc.values, color="blue", alpha=0.5)
ax[0,2].set_title("Z Accelerometro")
ax[0,2].plot(csvMeanSensorsSit.ZAcc.values, color="green", alpha=0.5)
ax[0,2].plot(csvMeanSensorsWalk.ZAcc.values, color="blue", alpha=0.5)
ax[1,0].set_title("X Giroscopio")
ax[1,0].plot(csvMeanSensorsSit.XGyro.values, color="green", alpha=0.5)
ax[1,0].plot(csvMeanSensorsWalk.XGyro.values, color="blue", alpha=0.5)
ax[1,1].set_title("Y Giroscopio")
ax[1,1].plot(csvMeanSensorsSit.YGyro.values, color="green", alpha=0.5)
ax[1,1].plot(csvMeanSensorsWalk.YGyro.values, color="blue", alpha=0.5)
ax[1,2].set_title("Z Giroscopio")
ax[1,2].plot(csvMeanSensorsSit.ZGyro.values, color="green", alpha=0.5)
ax[1,2].plot(csvMeanSensorsWalk.ZGyro.values, color="blue", alpha=0.5)
ax[2,0].set_title("X Magnetometro")
ax[2,0].plot(csvMeanSensorsSit.XMagn.values, color="green", alpha=0.5)
ax[2,0].plot(csvMeanSensorsWalk.XMagn.values, color="blue", alpha=0.5)
ax[2,1].set_title("Y Magnetometro")
ax[2,1].plot(csvMeanSensorsSit.YMagn.values, color="green", alpha=0.5)
ax[2,1].plot(csvMeanSensorsWalk.YMagn.values, color="blue", alpha=0.5)
ax[2,2].set_title("Z Magnetometro")
ax[2,2].plot(csvMeanSensorsSit.ZMagn.values, color="green", alpha=0.5, label="Sit")
ax[2,2].plot(csvMeanSensorsWalk.ZMagn.values, color="blue", alpha=0.5, label="Walk")

plt.legend(loc="lower right", shadow=True,borderpad=2)
plt.show()
```
