# Grafici Media Sit e Walk
In questo script viene usato il file di output "MeanSensorsSit.csv" e "MeanSensorsWalk.csv" ricavati dallo script 03_mean_sit_walk per graficare i dati usati per addestrare la rete neurale MLP.

```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from os.path import join
```

Viene specificato il percorso della cartella principale contenenti i file "MeanSensorsSit.csv" e "MeanSensorsWalk.csv".

```python
mainFilePath = "C:\\Users\\angel\\OneDrive\\Desktop\\progettoBiometria\\hmog_dataset\\public_dataset"
meanSensorsSitPath=join(mainFilePath,"MeanSensorsSit.csv")
meanSensorsWalkPath=join(mainFilePath,"MeanSensorsWalk.csv")
meanSensorsSit=pd.read_csv(meanSensorsSitPath)
meanSensorsWalk=pd.read_csv(meanSensorsWalkPath)
```

Vengono letti da "meanSensorsSit" e "meanSensorsWalk" le cordinate X, Y e Z di accelerometro, giroscopio e magnetometro. 
Vengono mostrati i grafici per X,Y,Z dei rispettivi sensori. 
Nell'ultimo grafico (per una questione di visibilit√† migliore) viene rappresentata la legenda dove il colore verde rappresenta i dati per Sit, mentre il colore blu rappresenta i dati per Walk.

```python
fig, plots = plt.subplots(nrows=3, ncols=3,figsize=(15,15))
for i in range(0,3):
    for j in range(0,3):
        col=(i*3)+j
        sitVal=meanSensorsSit.iloc[:,[(col+1)]]
        walkVal=meanSensorsWalk.iloc[:,[(col+1)]]
        name=sitVal.columns[0]
        p=plots[i,j]
        p.set_title(name)
        p.set_ylabel("Value")
        p.plot(sitVal, color="green", alpha=0.5, label="Sit")
        p.plot(walkVal, color="blue", alpha=0.5, label="Walk")
       
p=plots[2,2]
p.legend(loc="lower right", shadow=True,borderpad=2)
plt.show()
```
